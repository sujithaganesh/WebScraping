```python
soup=BeautifulSoup(page.content,'html')
```

```python
soup
```

```python
week=soup.find(id='globalContainer')
```

```python
print(week)
```

```python
items=week.find_all('div',class_='_5pcr userContentWrapper')
```

```python
#print(items)
len(items)
```

```python
print(items[0].find(class_='timestampContent').get_text())
print(items[0].find(class_='_5pbx userContent _3576').get_text())
print(items[0].find(class_='_51mq img sp_daR8qCHGLan sx_4d7322'))
```

```python
time=[item.find(class_='timestampContent').get_text() for item in items]
rev=[item.find(class_='_5pbx userContent _3576').get_text() for item in items]
rat=[item.find(class_='_51mq img sp_daR8qCHGLan sx_4d7322') for item in items]
```

```python
print(time)
print(rev)
print(rat)
```
